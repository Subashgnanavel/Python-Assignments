```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.__marks = marks

    def get_marks(self):
        return self.__marks

    def set_marks(self, marks):
        if 0 <= marks <= 100:
            self.__marks = marks
        else:
            print("Invalid marks! Must be between 0 and 100.")


s = Student("Subash", 85)
print("Name:", s.name)
print("Marks:", s.get_marks())
s.set_marks(92)
print("Updated Marks:", s.get_marks())
```

```python
class Vehicle:
    def start(self):
        print("Vehicle is starting...")

class Car(Vehicle):
    def drive(self):
        print("Car is driving...")


my_car = Car()
my_car.start()
my_car.drive()
```
