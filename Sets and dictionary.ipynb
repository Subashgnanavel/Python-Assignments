Create a dictionary with default values


```python
keys = ['subash', 'gnanavel', 'kgm']
values = 'data'
a_dict = dict.fromkeys(keys, values)
print(a_dict)
```

use a for loop to print the keys and values from a dictionary


```python
keys = ['subash', 'gnanavel', 'kgm']
values = 'data'
my_dict = dict.fromkeys(keys, values)
for i in my_dict:
    print(i,my_dict[i])
```
